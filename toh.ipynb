```python
def tower_of_hanoi(n, A, C, B):  
    if n == 1:
        print(f"Move disk 1 from {A} to {C}")
        return

    tower_of_hanoi(n - 1, A, B, C)  
    print(f"Move disk {n} from {A} to {C}")  
    tower_of_hanoi(n - 1, B, C, A)  


tower_of_hanoi(10, 'A', 'C', 'B')
```

```python
class ArrayADT:
    def __init__(self, capacity):
        self.arr = [None] * capacity
        self.size = 0

    def insert(self, index, value):
        if self.size >= len(self.arr):
            raise Exception("Array full")
        for i in range(self.size, index, -1):
            self.arr[i] = self.arr[i-1]
        self.arr[index] = value
        self.size += 1

    def delete(self, index):
        for i in range(index, self.size - 1):
            self.arr[i] = self.arr[i+1]
        self.arr[self.size - 1] = None
        self.size -= 1

    def get(self, index):
        return self.arr[index]
```

```python
a = ArrayADT(5)
a.insert(0, 10)
a.insert(1, 20)
a.insert(2, 30)
a.insert(1, 15)

print("Array after insertions:", a.arr[:a.size])  

a.delete(2)  
print("Array after deletion:", a.arr[:a.size])  

print("Element at index 1:", a.get(1)) 

```

```python
class node:
    def __init__(self,data):
        self.d=data
        self.next=None

class linkedList:
    def __init__(self):
        self.head=None

    def append(self,data):
        dNode=node(data)
        if self.head==None:
            self.head=dNode
        else:
            t=self.head
            while(t.next):
                t=t.next
            t.next=dNode

    def display(self):
        t=self.head
        while(t):
            print(t.d, end="->")
            t=t.next

l1=linkedList()
l1.append(10)
l1.append(20)
l1.append(30)
l1.display()
```
